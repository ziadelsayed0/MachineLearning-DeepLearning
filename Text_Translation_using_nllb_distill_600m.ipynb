```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("translation", model="facebook/nllb-200-distilled-600M")
```

```python
user_input = ''''
always you will need to have a good relationship with ALLAh
'''''
```

```python
source = 'eng_Latn'
target = 'arz_Arab'
translated_text = pipe(user_input, src_lang=source,tgt_lang=target)
print(translated_text)
```
